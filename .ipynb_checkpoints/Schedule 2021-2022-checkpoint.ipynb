```python
import pandas as pd
```

```python
table_sched = pd.read_html('https://www.whostheref.com/data-feed.php?feed_id=24698712&feed_format=table&dbtu=rugby&date=future&show_time=1&title_row=2&single_apps_col=1')
```

```python
df = table_sched[0]
df.columns=['Date', 'Time', 'Competition', 'Home', 'Away', 'Officials']
```

```python
df_huns = df.loc[df['Home'].str.contains('Huns') | df['Away'].str.contains('Huns')]
```

```python
df_huns1 = df_huns.loc[df_huns['Competition'].str.contains('D1')]
df_huns2 = df_huns.loc[df_huns['Competition'].str.contains('D2')]
df_huns3 = df_huns.loc[df_huns['Competition'].str.contains('D3')]
```
