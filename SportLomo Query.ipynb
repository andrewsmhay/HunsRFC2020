## Build the request

```python
req = requests.post('https://usarugby.sportlomo.com/wp-admin/admin-ajax.php?action=getClubMembers',
    headers = theheaders,
    data=teamid
)
```

# Build the NumPy array of results

```python
soup = BeautifulSoup(req.text)
header_array = []
tbody = soup.find("tbody")
td_list = tbody.find_all("td")

itlist = []
for i in td_list:
    itlist.append(i.text)
itlist = np.asarray(itlist)
itlist = np.split(itlist, len(itlist)/7)
```

# Build the data frame

```python
df = pd.DataFrame(itlist, columns = ['Region','Club','Number','First Name','Last Name','Category','Status'])
print('Rows:',len(df))
```

## Remove duplicates

```python
#df.duplicated(subset=['First Name', 'Last Name'])
df = df.drop_duplicates(subset=['First Name', 'Last Name'])
print('Rows:',len(df))
```
